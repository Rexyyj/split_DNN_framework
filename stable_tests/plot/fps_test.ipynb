```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
colors = ["tab:blue", "tab:orange","tab:green","tab:purple","tab:brown","tab:pink"]
```

```python
map_df_5fps = pd.read_csv("../measurements/sensitivity_5_fps/map.csv")
map_df_10fps = pd.read_csv("../measurements/sensitivity_10_fps/map.csv")
map_df_15fps = pd.read_csv("../measurements/sensitivity_15_fps/map.csv")
map_df_30fps = pd.read_csv("../measurements/sensitivity_30_fps/map.csv")

cha_df_5fps = pd.read_csv("../measurements/sensitivity_5_fps/characteristic.csv")
cha_df_10fps = pd.read_csv("../measurements/sensitivity_10_fps/characteristic.csv")
cha_df_15fps = pd.read_csv("../measurements/sensitivity_15_fps/characteristic.csv")
cha_df_30fps = pd.read_csv("../measurements/sensitivity_30_fps/characteristic.csv")
```

```python
cha_df_5fps.head()
```

```python
no_split_map_30 = 0.524
no_split_map_15 = 0.525
no_split_map_10 = 0.542
no_split_map_5 = 0.550
```

```python
plt.figure(figsize=(6,4))
map_df_group = map_df_5fps.groupby("pruning_thresh")
cha_df_group = cha_df_5fps.groupby("pruning_thresh")
keys = list(map_df_group.groups.keys())
for i in range(len(keys)):
    key = keys[i]
    map_df = map_df_group.get_group(key)
    maps = map_df["map"].to_numpy()*100
    sens = map_df["sensitivity"].to_numpy()*100

    cha_df = cha_df_group.get_group(key)
    cha_df_g = cha_df.groupby("quality")
    datasize = cha_df_g.mean()["datasize_est"].to_numpy()/1000

    # plt.plot(datasize,maps, marker="." ,color=colors[i],label="map, thr= "+str(key))
    plt.plot(datasize,sens, marker="^" ,color=colors[i],label="thr= "+str(key))
# plt.scatter(0,no_split_map_5*100,60,marker="*",color="tab:pink",label="No split")
plt.legend()
plt.grid()
plt.ylim([60,85])
plt.ylabel("Detection Sensitivity [%]")
plt.xlabel("Transferred data size [KB]")
```
