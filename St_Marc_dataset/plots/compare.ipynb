```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_regression = pd.read_csv("../measurements/regression/30_fps/map.csv")
df_jpeg = pd.read_csv("../measurements/jpeg/30_fps/map.csv")
```

```python
no_split_map_30 = 0.524
```

```python
plt.figure(figsize=(6,4))

jpeg_group = df_jpeg.groupby("pruning_thresh")
regression_group = df_regression.groupby("pruning_thresh")
key =0.1
jpeg_df = jpeg_group.get_group(key)
regression_df = regression_group.get_group(key)

plt.plot(jpeg_df["data_size_mean"].to_numpy()/1000,jpeg_df["map"].to_numpy()*100, marker="." ,label="jpeg")
plt.plot(regression_df["data_size_mean"].to_numpy()/1000/8,regression_df["map"].to_numpy()*100, marker="." ,label="regression")
plt.scatter(0,no_split_map_30*100,60,marker="*",color="tab:pink",label="No split")

plt.grid()
plt.legend()
plt.xlabel("Transferred data size [KB]")
plt.ylabel("DNN Detection mAP [%]")
```
